```python
import numpy as np
import pandas as pd

from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense,Embedding,LSTM,SpatialDropout1D
import re
```

```python
data=pd.read_excel('train.xlsx')
```

```python
data['text']=data['SentimentText']
```

```python
data.head()
```

```python
data['text']=data['text'].apply(lambda x:str(x).lower())
data['text']=data['text'].apply((lambda x:re.sub('[^a-zA-Z0-9\s]',"",x)))
print(data[ data['Sentiment'] == 1].size,data[ data['Sentiment'] == 0].size)
```

```python
tokenizer=Tokenizer(num_words=50000,split=' ')
tokenizer.fit_on_texts(data['text'].values)
X=pad_sequences(tokenizer.texts_to_sequences(data['text'].values),maxlen=60)
```

```python
embed_dim=64
lstm_out=100
model=Sequential()
model.add(Embedding(50000,embed_dim,input_length=60))
model.add(SpatialDropout1D(.4))
model.add(LSTM(lstm_out))
model.add(Dense(2,activation='sigmoid'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['acc'])
print(model.summary())
```

```python
Y=pd.get_dummies(data['Sentiment']).values
```

```python
len(X),len(Y)
```

```python
X.shape,Y.shape
```

```python
model.fit(X,Y,epochs=2,batch_size=32)
```

```python
model.save('model.h5')
```
